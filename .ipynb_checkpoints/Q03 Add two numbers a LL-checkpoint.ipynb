```python
def addTwoNumbers(head1: Node, head2: Node) -> Node:
    # Write your code here.
    
    if head1 == None:
        return head2
    elif head2 == None:
        return head1
    
    dummy = Node(0)
    nd = dummy

    carryOver = 0

    while head1 != None and head2 != None:
        
        addn = head1.data + head2.data
        total = addn + carryOver
        nodeVal = total % 10
        carryOver = total // 10

        newNode = Node(nodeVal)
        nd.next = newNode
        nd = newNode

        head1 = head1.next
        head2 = head2.next

    if head1 == None:
        while head2 != None:
            addn = head2.data + carryOver
            nodeVal = addn % 10
            carryOver = addn // 10
            
            newNode = Node(nodeVal)
            nd.next = newNode
            nd = newNode
            head2 = head2.next
    elif head2 == None:
        while head1 != None:
            addn = head1.data + carryOver
            nodeVal = addn % 10
            carryOver = addn // 10
            
            newNode = Node(nodeVal)
            nd.next = newNode
            nd = newNode
            head1 = head1.next
    
    if carryOver != 0:
        newNode = Node(carryOver)
        nd.next = newNode
        
    return dummy.next
    
```

```python
def addTwoNumbers(head1: Node, head2: Node) -> Node:
    # Write your code here.
    carry = 0

    dummy = Node(0)
    temp = dummy

    while head1 or head2 or carry:

        sums = 0

        if head1 != None:
            sums += head1.data
            head1 = head1.next
        
        if head2 != None:
            sums += head2.data
            head2 = head2.next
        
        sums += carry

        nodeVal = sums % 10
        carry = sums // 10

        temp.next = Node(nodeVal)
        temp = temp.next

    return dummy.next
```
